```python
# Load the data from SQLite database
conn = sqlite3.connect("ukr_aid.db")

# SQL query to join the 'commitments', 'allocations', and 'countries' tables
query = """
SELECT 
    co.country_name as Country, 
    c.gdp_2021 as 'GDP in 2021($ billion)', 
    c.financial_commitments as 'Financial commitments($ billion)', 
    c.humanitarian_commitments as 'Humanitarian commitments($ billion)', 
    c.military_commitments as 'Military commitments($ billion)',
    a.financial_allocations as 'Financial allocations($ billion)', 
    a.humanitarian_allocations as 'Humanitarian allocations($ billion)', 
    a.military_allocations as 'Military allocations($ billion)'
FROM commitments c
JOIN allocations a ON c.country_id = a.country_id
JOIN countries co ON c.country_id = co.country_id
"""

# Execute the query and load data into a DataFrame
df = pd.read_sql(query, conn)
conn.close()

# Ensure that 'df' is defined and not empty
if df.empty:
    print("Data not loaded correctly.")
else:
    print("Data loaded successfully.")
    print(df.head())

# Check for missing values and handle them (fill with 0 in this case)
df = df.fillna(0)

# Correlation Matrix (using only numeric columns)
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import mlflow
import mlflow.sklearn
from sklearn.model_selection import cross_val_score
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, confusion_matrix

# Set up DAGsHub authentication
os.environ['MLFLOW_TRACKING_USERNAME'] = 'HarshithReddy-Audipudi'
os.environ['MLFLOW_TRACKING_PASSWORD'] = 'Qwerty@123'

# Set up MLflow
mlflow.set_tracking_uri("https://dagshub.com/HarshithReddy-Audipudi/ML-project-haudipud.mlflow")
mlflow.set_experiment("Experiment_1_Logistic_Regression")

# Assuming X_train, X_test, y_train, y_test are already defined

# Define preprocessing steps
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object', 'bool']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ]), numeric_features),
        ('cat', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ]), categorical_features)
    ])

# Create pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

with mlflow.start_run():
    # Perform cross-validation
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=10, scoring='f1')
    
    # Fit the model on the entire training data
    pipeline.fit(X_train, y_train)
    
    # Make predictions on test data
    y_pred = pipeline.predict(X_test)
    
    # Calculate F1 score and confusion matrix
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    # Log metrics
    mlflow.log_metric("mean_cv_f1", np.mean(cv_scores))
    mlflow.log_metric("std_cv_f1", np.std(cv_scores))
    mlflow.log_metric("test_f1", f1)
    
    # Log model
    mlflow.sklearn.log_model(pipeline, "logistic_regression_model")

print(f"Mean CV F1-score: {np.mean(cv_scores):.4f} (+/- {np.std(cv_scores):.4f})")
print(f"Test F1-score: {f1:.4f}")
print("Confusion Matrix:")
print(cm)
```
